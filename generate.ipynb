```python
import toolz as tz
import random

is_empty = lambda s: s.strip() == ""

def is_valid_rule(lines): 
    start = lines[0].strip()
    return start.startswith('#') and start.endswith('->')

def parse_rule(lines):
    key = lines[0].strip().rstrip("->").strip()
    values = map(str.strip, lines[1:])
    return (key,values)

def has_terminated(words):
    return "#" not in "".join(words)

def generate_sentence(rules): 
    def replace_word(word): 
        if not word.startswith("#"):
            return word
        else: 
            return random.choice(rules[word])    
    
    words = random.choice(rules['#Start']).split(" ")
    
    for i in range(10): 
        words = map(replace_word, words)
        if has_terminated(words):
            break
            
    return " ".join(words)

grammar = open('./grammar-2.txt').read()

my_rules = tz.thread_last(
    grammar.split('\n'),
    (tz.partitionby, is_empty),
    (filter, is_valid_rule),
    (map, parse_rule),
    dict
)

n = 100
print "\n".join([generate_sentence(my_rules) for _ in range(n)])

# dict(my_rules)
```
